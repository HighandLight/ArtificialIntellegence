```python
import numpy as np
import matplotlib.pyplot as plt # 그림 그리는 라이브러리
```

```python
x1 = np.array([2, 4, 6, 8])
x2 = np.array([2,2,2,2]) # 문제집 푼 개수
y = np.array([55, 85, 70, 95])

# 인간 : a:5, b:50
a1 = 0
a2 = 0
b = 0

for i in range(1, 5001):
  y_hat = a1 * x1 + a2 * x2 + b
  error = y_hat - y
  
  a1_diff = sum(2 * error * x1)
  a2_diff = sum(2 * error * x2)
  b_diff = sum(2 * error * 1)

  a1 = a1 - a1_diff * 0.005
  a2 = a2 - a2_diff * 0.005
  b = b - b_diff * 0.005

  print(f'{i}회 학습, 기울기1 {a1}, 기울기2 {a2}, 절편 {b}')
```

```python
plt.scatter(x1, y)
plt.plot(x1, a1*x1 + a2*x2 + b, 'r')
```

```python
money = 2000
card = True

if money >= 3000 and card == True:
  print('taxi')
else:
  print('walk')
```

```python
pocket = ['paper', 'cellphone']
card = True

if 'money' in pocket:
  print('taxi')
else:
  if card == True:
    print('taxi')
  else:
    print('walk')
```

```python
pocket = ['paper', 'cellphone']
card = True

if 'money' in pocket:
  print('taxi')
elif card == True:
  print('taxi')
else:
  print('walk')
```

```python
# 반복문

# for : 순서가 있는 유한반복

box = ['곽경일', '장원영', '아이유', '백현']

for i in box: # box 안에 있는 것을 순서대로 i라는 변수에 넣어라
  print(i)
```

```python
a = [90, 25, 75, 45, 100]
num = 0

for i in a:
  num = num + 1
  if i > 60:
    print(num, '번 학생 합격', i, '점')
```
